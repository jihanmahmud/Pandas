```python
delivery=pd.read_csv('deliveries.csv')
delivery.head()
```

```python
delivery.shape
```

```python
delivery.groupby('batsman')
```

```python
run=delivery.groupby('batsman')
run['batsman_runs'].sum().sort_values(ascending=False)
```

```python
mask=delivery['batsman_runs']==6
new_delivery=delivery[mask]
```

```python
new_delivery.shape
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head()
```

```python
new_delivery.groupby('bowler')['batsman_runs'].count().sort_values(ascending=False).head()
```

```python
dw=delivery[delivery['batsman']=='DA Warner']
dw.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def run_scored(batsman_name):
    dw=delivery[delivery['batsman']==batsman_name]
    return dw.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
run_scored('DA Warner')
```
