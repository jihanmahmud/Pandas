```python
dw=delivery [delivery['batsman']=='DA warner']
dw.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def run_scored(batsman_name):
    dw=delivery[delivery['batsman']==batsman_name]
    return dw.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
run_scored('DA Warner')
```

## ISIN functions

```python
delivery.head(3)
```

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x=all_batsman>200
x
```

```python
all_batsman[x].head(3)
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
batsman_list
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
runs=final.groupby('batsman')['batsman_runs'].sum()
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
sr.sort_values(ascending=False)
```
